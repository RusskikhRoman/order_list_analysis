```python
# 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)?
# Не стоит забывать, что внутри месяца может быть не целое количество недель.
# Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть.
```

```python
df = (
olist_orders_dataset
    .groupby(['customer_unique_id', 'month_and_year'])
    .agg({'order_id': 'count'})
    .sort_values('order_id', ascending = False)
    .reset_index()
)
df['m_n_y'] = pd.to_datetime(df['month_and_year'], format = '%B, %Y')
df['year'] = df['m_n_y'].dt.year
df['month'] = df['m_n_y'].dt.month
df['num_of_weeks'] = (df[['year', 'month']].apply(lambda x: monthrange(x['year'], x['month'])[1], axis = 1))/7
df = df.drop(columns=['m_n_y', 'year', 'month'])
df['order_id'] = df['order_id']/df['num_of_weeks']
df = df.rename(columns={'order_id': 'purchases_per_week'})
df = df.drop('num_of_weeks', axis = 1)
df
```

```python
# 5.Используя pandas, проведи когортный анализ пользователей.
# В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.
```

```python
df_cohort = olist_orders_dataset
```

```python
# создаем столбец, отображающий месяц первой покупки пользователя
df_cohort['cohort_month'] = df_cohort.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')

# определяем lifetime для каждой покупки в месяцах с момента первой покупки
df_cohort['lifetime'] = (df_cohort['order_purchase_timestamp'] - df_cohort['cohort_month'].dt.to_timestamp()).dt.days // 30

# группируем пользователей по когортам и lifetime
cohorts = df_cohort.groupby(['cohort_month', 'lifetime']).agg({'customer_unique_id': pd.Series.nunique}).reset_index()

# создаем сводную таблицу, где на пересечении месяца когорты и lifetime отображается количество уникальных пользователей
cohorts_pivot = cohorts.pivot_table(index='cohort_month', columns='lifetime', values='customer_unique_id')

# определяем размер каждой когорты
cohort_sizes = cohorts_pivot.iloc[:, 0]

# вычисляем retention rate для каждой когорты и lifetime
retention = cohorts_pivot.divide(cohort_sizes, axis=0)
```

```python
# выделяем когорты, образованные в 2017 году
cohorts_2017 = retention.loc['2017']

# находим retention rate на 3-й месяц для каждой когорты
retention_3rd_month = cohorts_2017.iloc[:, 3]

# находим когорту с максимальным retention rate на 3-й месяц
highest_retention_cohort = retention_3rd_month.idxmax()

# выводим результаты
print(f'Когорта с самым высоким retention rate на 3-й месяц: {highest_retention_cohort}')
```
